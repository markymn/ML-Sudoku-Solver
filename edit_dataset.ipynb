```python
# Create new folder for the editted dataset
```

```python
new_path = "C:/Users/markn/PycharmProjects/ML-Sudoku-Solver/editted_dataset"
if not os.path.exists(new_path):
    os.makedirs(new_path)
```

```python
# Resize and invert colors of images and save them in a new folder, the editted dataset
```

```python
num = 0

for folder in os.listdir(path):

    count = 0

    subfolder_path = new_path + "/" + str(num)
    if not os.path.exists(subfolder_path):
        os.makedirs(subfolder_path)
    
    for img_file in os.listdir(path + "/" + folder):
        
        img = Image.open(path + "/" + folder + "/" + img_file)
        img = img.resize((28, 28))
        img = ImageOps.invert(img)
        img = img.convert('1')
        img.save(subfolder_path + "/" + str(num) + "-" + str(count) + ".png")

        count += 1

    num += 1 
```
